## Análise exploratória dos dados

```python
# Visão geral dos dados com describe
top_2017.describe()
```

```python
# Visão geral dos dados com info
top_2017.info()
```

```python
# Selecionar apenas colunas numéricas
numerical_columns = top_2017.select_dtypes(include=['float64', 'int64'])

# Gerar uma análise descritiva para todas as variáveis numéricas do conjunto de dados
description = numerical_columns.describe()

# Analisar a quantidade de valores ausentes (NAs) por coluna
missing_values = top_2017.isna().sum()

# Verificar a presença de outliers utilizando o método do IQR (Interquartile Range)
Q1 = numerical_columns.quantile(0.25)
Q3 = numerical_columns.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numerical_columns < (Q1 - 1.5 * IQR)) | (numerical_columns > (Q3 + 1.5 * IQR))).sum()

# Gerar uma lista com as informações coletadas para cada variável numérica
summary = pd.DataFrame({
    'Missing Values': missing_values,
    'Outliers': outliers,
    'Min Value': description.loc['min'],
    'Max Value': description.loc['max'],
    'Mean': description.loc['mean'],
    # Variância calculada diretamente como variância amostral
    'Variance': description.loc['std']**2,
    'Standard Deviation': description.loc['std'],
})

# Exibir a tabela resumo
summary
```

```python
# Definição do diretório de figures
figures_dir = Path('../figures')
os.makedirs(figures_dir, exist_ok=True)

# Calcular a matriz de correlação para as colunas numéricas
corr_matrix = top_2017.select_dtypes(include=['float64', 'int64']).corr()

# Criar um mapa de calor para a matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matriz de Correlação das Variáveis Numéricas - Spotify Top Tracks 2017")
plt.tight_layout()
plt.savefig(os.path.join(figures_dir, 'correlation_matrix.png'))
plt.close()

# Selecionar algumas variáveis para visualização
selected_columns = ['danceability', 'energy', 'acousticness', 'valence', 'duration_ms', 'tempo']

# Plotar gráficos de dispersão para variáveis com correlações interessantes
pairplot = sns.pairplot(top_2017[selected_columns])
pairplot.fig.suptitle("Scatterplots das Variáveis Selecionadas - Spotify Top Tracks 2017", y=1.02)
pairplot.savefig(os.path.join(figures_dir, 'pairplot.png'))
plt.close()
```
